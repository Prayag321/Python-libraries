```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python


def main():
	# Data for plotting
	x = [0, 1, 2, 3, 4, 5]
	y = [0, 1, 4, 9, 16, 25]

	# Plot the line
	plt.plot(x, y, label='y = x^2')

	# Add labels to the x and y axes
	plt.xlabel('X Axis Label')
	plt.ylabel('Y Axis Label')

	# Add a title to the plot
	plt.title('Line Plot Example')

	# Add a legend to the plot
	plt.legend() #That is label'y = x^2'

	# Display the plot
	plt.show()

 
if __name__ == "__main__":
	main()
```

## 2. Write a Python program to draw a line using given axis values with suitable label in the x axis, y axis and a title

```python

def main():
	# Data for plotting
	x = ['PRAYAG','DEVEN',]
	y = [95,90]

	# Plot the line
	plt.plot(x, y, label='MARKS', marker='*', linestyle='--', color='b')
	plt.ylim(0, 100)

	# Add labels to the x and y axes
	plt.xlabel('X Axis Label')
	plt.ylabel('Y Axis Label')

	# Add a title to the plot
	plt.title('Line Plot Example')

	# Add a legend to the plot
	plt.legend() #That is label'y = x^2'

	# Display the plot
	plt.show()
 
if __name__ == "__main__":
	main()

```

## 3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
Test Data:<br>
test.txt<br>
1 2<br>
2 4<br>
3 1


```python
def main():
	data = np.loadtxt('text.txt')

	x = data[:,0]
	y = data[:,1]

	plt.plot(x,y, marker='*',ls="--",c='g')
	plt.title("Line plot",loc='left')#explore default center

	plt.xlabel('X Axis ')
	plt.ylabel('Y Axis ')

	plt.show()
 
if __name__ == "__main__":
	main()
```

## 4. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.

```python
def main():
	df = pd.read_csv("test.csv")
	print(df)

	plt.figure(figsize=(10,10))

	x_date = df.loc[:,'Date'].tolist()
	y_open= df['Open'].tolist()


	plt.subplot(2,2,1)
	plt.plot(x_date,y_open,label='open',c='b')
	plt.legend()


	plt.subplot(2,2,2)
	plt.plot(x_date,df["High"].tolist(),label='High',c="g")
	plt.legend()


	plt.subplot(2,2,3)
	plt.plot(x_date,df["Low"].tolist(),label='Low',c='r')
	plt.legend()


	plt.subplot(2,2,4)
	plt.plot(x_date,df["Close"].tolist(),label='Close')
	plt.legend()

	plt.tight_layout()
	plt.show()	
if __name__ == "__main__":
	main()
```
