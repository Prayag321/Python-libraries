```python
import logging
import plotly.express as px
import pandas as pd
import numpy as np
import plotly.io as pio
import matplotlib.pyplot as plt
```

## 1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python


def main():
    # configure logging
    logging.basicConfig(
        filename="plotly_logs.log", 
        level=logging.INFO, 
        format='%(asctime)s:%(levelname)s:%(message)s'
    )
    
    n = 1000
    x = np.random.rand(n)
    y = np.random.rand(n)
    
    fig = px.scatter(x=x, y=y, title='Scatter Plot of Random Coordinates')
    pio.renderers.default = 'browser' #to show in browser
    fig.show()
    logging.info("Plotly Q1:Scatter plot with 1000 random coordinates was generated and displayed.")

if __name__ == "__main__":
    main()
```
