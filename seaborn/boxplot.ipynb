```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```


## 1. Write a program to draw box plot of life expectancy by continent for a data set given in a url
https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapmind
er-FiveYearData.csv

```python
def main():
  url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
  df = pd.read_csv(url)
  print(df.head(4))

  # sns.barplot(x=df['sex'],y=df['survived'])
  sns.boxplot(data=df,x='lifeExp',y='continent', color='cyan') #easy method


if __name__=="__main__":
  main()
```

## 2. Write a program to draw a box plot of day by tips for a dataset given in a url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
def main():

  df = sns.load_dataset('tips')
  print(df.head())
  # sns.barplot(x=df['sex'],y=df['survived'])
  sns.boxplot(data=df,x='day',y='tip',hue='sex',gap=.2 ,hue_order=['Female','Male'],fliersize=10) #easy method


if __name__=="__main__":
  main()
```
