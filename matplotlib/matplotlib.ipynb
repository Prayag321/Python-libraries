## 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python


def main():
	# Data for plotting
	x = [0, 1, 2, 3, 4, 5]
	y = [0, 1, 4, 9, 16, 25]

	# Plot the line
	plt.plot(x, y, label='y = x^2')

	# Add labels to the x and y axes
	plt.xlabel('X Axis Label')
	plt.ylabel('Y Axis Label')

	# Add a title to the plot
	plt.title('Line Plot Example')

	# Add a legend to the plot
	plt.legend() #That is label'y = x^2'

	# Display the plot
	plt.show()

 
if __name__ == "__main__":
	main()
```

## 3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.

## 5. Write a Python program to plot two or more lines on same plot with suitable legends of each line

```python
def main():
	df = pd.read_csv("test.csv")
	print(df)

	plt.figure(figsize=(10,10))

	x_date = df.loc[:,'Date'].tolist()
	y_open= df['Open'].tolist()



	plt.plot(x_date, y_open, label='Open', c='b', marker="D")
	plt.plot(x_date, df["High"].tolist(), label='High', c="g", marker="s")
	plt.plot(x_date, df["Low"].tolist(), label='Low', c='r', marker="H")
	plt.plot(x_date, df["Close"].tolist(), label='Close', marker="*")
	plt.legend()

	plt.show()	
if __name__ == "__main__":
	main()
```

## 12. Write a Python program to create multiple types of charts

```python

x = ['A', 'B', 'C', 'D']
y = [10, 20, 15, 30]

# Line Chart
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)  # (rows, columns, panel number)
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Line Chart')
plt.xlabel('x')
plt.ylabel('y')

# Bar Chart
plt.subplot(2, 2, 2)
plt.bar(x, y, color='green')
plt.title('Bar Chart')
plt.xlabel('x')
plt.ylabel('y')

# Scatter Plot
x = np.random.rand(10) * 10
y = np.random.rand(10) * 10
sizes = np.random.randint(50, 200, size=10)  # Bubble sizes

plt.subplot(2, 2, 3)
plt.scatter(x, y, s=sizes, alpha=0.5, c='red')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Pie Chart
pie_sizes = [30, 25, 20, 25]
labels = ['Slice A', 'Slice B', 'Slice C', 'Slice D']

plt.subplot(2, 2, 4)
plt.pie(pie_sizes, labels=labels, autopct='%1.1f%%', colors=['r','g','b','#ff00ff'])
plt.title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Display the charts
plt.show()
```

## 14. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks.  

```python
dates = ['03-10-16', '04-10-16', '05-10-16', '06-10-16', '07-10-16']
closing_values = [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]

plt.figure(figsize=(10, 6))
plt.plot(dates, closing_values, marker='o', linestyle='-', color='b')

plt.grid(which='major', linestyle='-', linewidth=0.5, color='blue')  # Major grid
plt.grid(which='minor', linestyle=':', linewidth=0.5, color='blue')  # Minor grid
plt.minorticks_on()  # Turn on minor ticks for minor grids
plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False) #Turn of tics


plt.title('Alphabet Inc. Closing Values (Oct 3, 2016 to Oct 7, 2016)')
plt.xlabel('Date')
plt.ylabel('Closing Value')


plt.show()
```
