```python
import logging
import plotly.express as px
import pandas as pd
import numpy as np
import plotly.io as pio
import matplotlib.pyplot as plt
```

```python
# configure logging
logging.basicConfig(
  filename="plotly_logs.log", 
  level=logging.INFO, 
  format='%(asctime)s:%(levelname)s:%(message)s'
  )
  
```

## 1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
def main():
  n = 1000
  x = np.random.rand(n)
  y = np.random.rand(n)
  
  fig = px.scatter(x=x, y=y, title='Scatter Plot of Random Coordinates')
  pio.renderers.default = 'browser' #to show in browser
  fig.show()
  logging.info("Plotly Q1:Scatter plot with 1000 random coordinates was generated and displayed.")

if __name__ == "__main__":
  main()
```

## 2. Write a program to draw line and scatter plots for random 100 x and y coordinates

```python

def main():
  n = 100
  x = np.random.randint(0,n,n)
  y = np.random.randint(0,n,n)
  
  fig = px.scatter(x=x, y=y, title='Scatter Plot of Random Coordinates')
  fig.add_scatter(x=x, y=y,mode='lines', name='line')
  pio.renderers.default = 'browser' #to show in browser
  fig.show()
  logging.info("Plotly Q2:line and scatter plots for random 100 x and y coordinates was generated and displayed.")

if __name__ == "__main__":
  main()
```

## 3. Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python

def main():
  n = 500
  x = np.random.randint(0,n,n)
  y = np.random.randint(0,n,n)
  
  fig = px.scatter(x=x, y=y, title='Scatter Plot of Random Coordinates', template='plotly_dark')
  pio.renderers.default = 'browser' #to show in browser
  fig.show()
  logging.info("Plotly Q3:scatter plots for random 500 x and y coordinates with style was generated and displayed.")

if __name__ == "__main__":
  main()
```
